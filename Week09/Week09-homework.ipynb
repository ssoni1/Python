The data frame contains notes taken by staff members. Your task is to clean it up to make it usable for analysis.

<b><h2>Tasks</b></h2>
<br>
Create start date, end date columns and populate the values (If notes don't include month or day use 1) i.e: 2018 => 2018-01-01

Create phone number column (If notes don't include phone numbers, use None)

Create a column duration showing difference between start & end dates in years as a whole number

```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week09/data/hw_data.csv', index_col=0)
```

```python
df
```

```python
#Phone number regex pattern
pattern = re.compile(r'(\d{3})?((\(\d{3}\))|(\d{3})(\s|-?|.?))(\d{3}(\s|-?|.?))(\d{4})(\s?(([E|e]xt[:|.|]?))(\s?\d+))?')
```

```python
df['Phone_No']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
df.head(20)
```

```python
#Start Date Regex Pattern 
pattern = re.compile(r'((start|st\sdt|st\sdate|st\sDt|STDT)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4})',re.I)
```

```python
df['Start_Date']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
```

```python
#End Date Regex Pattern 
pattern = re.compile(r'(end|ed|en\sdt|endt)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4}',re.I)
```

```python
df['End_Date']= [pattern.search(i).group() if pattern.search(str(i)) != None else None  for i in df['notes'] ]
```

```python
df
```

```python
df['start_dte'] = df['Start_Date'].str.replace(r'(start|st\sdt|st\sdate|st\sDt|STDT)\D*', '')
```

```python
df['end_dte'] =  df['End_Date'].str.replace(r'(end|ed|en\sdt|ENDT)\D*', '')
```

```python
df.head(30)
```

```python
pattern = re.compile(r'(\b^[0-9]{2}(\/)[0-9]{2})$\b')
```

```python
df['Strt']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['start_dte']]
```

```python
df['End']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['end_dte']]
```

```python
df['Strt_Yr']=pd.DatetimeIndex(df['Strt']).year
df['End_Yr']=pd.DatetimeIndex(df['End']).year
```

```python
df.drop(['Start_Date','End_Date','start_dte','end_dte','notes'], axis=1, inplace=True)
df.head(3)
```

```python
df['Duration']=df['End_Yr']-df['Strt_Yr']
```
