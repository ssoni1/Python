## Working with Strings

The data frame contains notes taken by staff members. Your task is to clean it up to make it usable for analysis. 

### Tasks
* Create start date, end date columns and populate the values (If notes don't include month or day use 1) i.e: 2018 => 2018-01-01

* Create phone number column (If notes don't include phone numbers, use None)
* Create a column `duration` showing difference between start & end dates in years as a whole number

```python
import pandas as pd
import re 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week09/data/hw_data.csv', index_col=0)
```

```python
df.head(20)
```

```python
### creating a phone number column 

pattern = re.compile(r'(\d{3})?((\(\d{3}\))|(\d{3})(\s|-?|.?))(\d{3}(\s|-?|.?))(\d{4})(\s?(([E|e]xt[:|.|]?))(\s?\d+))?')
```

```python
df['Phone_no']= [pattern.search(s).group() if pattern.search(str(s)) != None else None  for s in df['notes'] ]
```

```python
df['Phone_no'] =  df['Phone_no'].str.replace(r'\D', '-')
```

```python
pattern = re.compile(r'(end|ed|en\sdt)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4}',re.I)
```

```python
df['End_dte']= [pattern.search(s).group() if pattern.search(str(s)) != None else None  for s in df['notes'] ]
```

```python
pattern = re.compile(r'((start|st\sdt|st\sdate|st\sDt|STDT)\D*\d{1,4}(\D)?\d{1,4}(\D)?\d{1,4})',re.I)
```

```python
df['Start_dte']= [pattern.search(s).group() if pattern.search(str(s)) != None else None  for s in df['notes'] ]
```

```python
df['Start_dte']= [pattern.search(s).group() if pattern.search(str(s)) != None else None  for s in df['notes'] ]
```

```python
df['Start_dte'] = df['Start_dte'].str.replace(r'(start|st\sdt|st\sdate|st\sDt|STDT)\D*', '')
```

```python
df['End_dte'] =  df['End_dte'].str.replace(r'(end|ed|en\sdt|ENDT)\D*', '')
```

```python
df.head()
```

```python
df['Start_dte'] =pd.to_datetime(df['Start_dte'],errors='coerce').dt.date
```

```python
df['End_dte'] =pd.to_datetime(df['End_dte'],errors='coerce').dt.date
```

```python
pattern = re.compile(r'(\b^[0-9]{2}(\/)[0-9]{2})$\b')
```

```python
df['Start_dte']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['Start_dte']]
```

```python
df['End_dte']=[pd.to_datetime(s,format='%m/%y').date() if pattern.search(str(s)) != None else pd.to_datetime(s,errors='coerce').date() for s in df['End_dte']]
```

```python
df.head(20)
```

```python
df['duration']=df['End_dte']-df['Start_dte']
```
